law of large numbers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

sample_sizes = 2**np.arange(4, 21)
sample_sizes
```

```python
normal_mean = [np.mean(np.random.standard_cauchy(size = i)) for i in sample_sizes ]
```

```python
plt.plot(np.arange(4,21), normal_mean, label='Mean') 
# add a horizontal line at y=0
plt.axhline(y=0, color='r', linestyle='-')
# add x and y labels
plt.xlabel('Sample Size')
plt.ylabel('Mean')
plt.show()
```

central limit theorem

```python
n = 100
Rep = 10**5

mu = 3
x = np.random.chisquare(mu, size = (n, Rep))
z = (x - mu) / np.sqrt(2*mu) # normalize to have mean 0 and variance 1

zbar = np.sqrt(n) * np.mean(z, axis = 0) # mean * factor
```

```python
from scipy.stats import norm

x_base = np.linspace(-4, 4, 100)
plt.plot(x_base, norm.pdf(x_base, loc = 0, scale = 1), 'r-', lw = 2, alpha = 0.6, label = 'norm pdf')
plt.hist(zbar, bins = 100, density = True)
plt.show()
```
